```python
import pandas as pd

df = pd.read_csv("used_cars.csv")
df.head()
```

```python
df.shape        # rows, columns
df.columns      # column names
df.info()       # data types & nulls
df.describe()   # statistics
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```

```python
df['Car_Age'] = 2025 - df['Year']
```

```python
df.columns
```

```python
df['Kilometers_Driven'] = df['Kilometers_Driven'].astype(int)
```

```python
df['Price'] = df['Price'].astype(float)
df['Year'] = df['Year'].astype(int)
df['Seats'] = df['Seats'].astype(int)
```

```python
df['Car_Age'] = 2025 - df['Year']
df[['Year', 'Car_Age']].head()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(6,4))
plt.hist(df['Price'], bins=30)
plt.xlabel("Car Price (Lakh)")
plt.ylabel("Count")
plt.title("Used Car Price Distribution")
plt.show()
```

```python
sns.scatterplot(x='Car_Age', y='Price', data=df)
plt.title("Car Age vs Price")
plt.show()
```

```python
sns.scatterplot(x='Kilometers_Driven', y='Price', data=df)
plt.title("Kilometers Driven vs Price")
plt.show()
```

```python
sns.boxplot(x='Fuel_Type', y='Price', data=df)
plt.title("Fuel Type vs Car Price")
plt.show()
```

```python
sns.boxplot(x='Transmission', y='Price', data=df)
plt.title("Transmission vs Price")
plt.show()
```

```python
sns.boxplot(x='Owner_Type', y='Price', data=df)
plt.title("Owner Type vs Price")
plt.show()
```

```python
sns.boxplot(x='Seats', y='Price', data=df)
plt.title("Seats vs Price")
plt.show()
```

```python
numeric_df = df[['Price', 'Kilometers_Driven', 'Car_Age', 'Seats']]
corr = numeric_df.corr()

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
